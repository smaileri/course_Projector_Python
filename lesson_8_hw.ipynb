1. Write a game where user should guess of a capital of the country that you have in your dictionary.

    You should show user a random country from the list and ask him to guess the capital. If user input right capital, print "You are right!", add him a point and ask him to guess another country. If not, you should ask again. User should be able to quit the game by typing "exit".
    You should print current score after each round. Also, you should print the final score before user quit the game.


    Optional: 
    1. Give user a hint if he guesses wrong. Hint should looks like first letter of the capital. If you user make another mistake, you should print one more letter from the capital.

    2. If user make a mistake you should decrement his lives. Initial amount of lives is 3. Game will end when user has no lives left. You should print the final score after user has no lives left.

```python
import random
capitals = {
        'Ukraine': 'Kyiv',
        'France': 'Paris',
        'Germany': 'Berlin',
        'Italy': 'Rome',
        'USA': 'Washington',
        'Canada': 'Ottawa',
        'Switzerland': 'Bern',
        'Austria': 'Vienna',
        'Belgium': 'Brussels',
        'Sweden': 'Stockholm',
        'Norway': 'Oslo',
        'Denmark': 'Copenhagen',
        'Finland': 'Helsinki',
        'Poland': 'Warsaw',
        'Romania': 'Bucharest',
        'Bulgaria': 'Sofia',
        'Greece': 'Athens',
    }
lives_total = 3
heart = "❤"
empty_heart = "♡"

def hint (key_country: str, pos: int) -> str:
    return capitals[key_country][:pos]

def print_hint (hint_text: str) -> None:
    print (f"The capital starts with: {hint_text}")

def print_lives (lives: int)-> None:
    print (f"You have {heart * lives}{empty_heart * (lives_total - lives)}")

def choose_random_country ()-> str:
    pos = random.randint(0, len(capitals) - 1)
    return list(capitals)[pos]

def ask_guess ( country: str) -> str:
    print((f"What is the capital of { country }? Be careful."))
    return input (f"What is the capital of { country }? Be careful.")

def print_score( score: int) -> None:
    print (f"You have {score} points.")

def play_round(lives) -> tuple: #(is_win: bool, lives :int, exited: bool):
    country = choose_random_country()
    position_hint = 0
    answered_correctly = False
    needs_hint = False

    while lives and (not answered_correctly):

        if needs_hint:
            position_hint += 1
            text = hint(country, position_hint)
            print_lives(lives)
            print_hint( text)
            
        data = ask_guess(country)

        if data == "exit":
            return (False, lives, True)
        elif data != capitals[country]:
            needs_hint = True
            lives -= 1
        else:
            answered_correctly = True
            # break
    
    return (answered_correctly, lives, False)
    


def game() -> None:
    score = 0
    lives = lives_total
    while lives:
        is_win, lives, exited = play_round (lives)
        if exited:
            print_lives ( lives)
            print_score ( score)
            print ("You exited.")
        if is_win:
            score += 1
            print_score( score)
            print ("You are right!")
    else:
        print_lives( lives)
        print_score( score)
        print ("The end!")

game()
        
        


```


2. **Optional** Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M.

    Symbol       Value
    I             1
    V             5
    X             10
    L             50
    C             100
    D             500
    M             1000

    For example, 2 is written as II in Roman numeral, just two ones added together. 12 is written as XII, which is simply X + II. The number 27 is written as XXVII, which is XX + V + II.

    Roman numerals are usually written largest to smallest from left to right. However, the numeral for four is not IIII. Instead, the number four is written as IV. Because the one is before the five we subtract it making four. The same principle applies to the number nine, which is written as IX. There are six instances where subtraction is used:

    * I can be placed before V (5) and X (10) to make 4 and 9. 
    * X can be placed before L (50) and C (100) to make 40 and 90. 
    * can be placed before D (500) and M (1000) to make 400 and 900.



```python
exceptions = {"IV" : "IIII",
            "IX" : "VIIII",
            "XL" : "XXXX",
            "XC" : "LXXXX",
            "CD" : "CCCC",
            "CM" : "DCCCC"
            }

roman_int = { "I" : 1,
            "V" : 5,
            "X" : 10,
            "L" : 50,
            "C" : 100,
            "D" : 500,
            "M" : 1000
            }

def make_pseudoroman_from_roman (s: str) ->str:
        for exceptions_case, pseudo_value in exceptions.items():
                s = s.replace (exceptions_case, pseudo_value)
        return s

def roman_to_int(s: str) -> int:
        s = make_pseudoroman_from_roman(s)
        answer = 0
        for char in s:
                answer += roman_int[char]
        return answer       

def test_roman_to_int():
        result1 = roman_to_int("III")
        assert result1 == 3

        result1 = roman_to_int("LVIII")
        assert result1 == 58

        result1 = roman_to_int("MCMXCIV")
        assert result1 == 1994 

test_roman_to_int()
```

4. Given an array nums of size n, return the majority element.

    The majority element is the element that appears more than any other element. You may assume that the majority element always exists in the array.

        Example 1:

        Input: nums = [3,2,3]
        Output: 3

        Example 2:

        Input: nums = [2,2,1,1,1,2,2]
        Output: 2

```python
from collections import defaultdict

def find_count(nums:list)->defaultdict:
    quantity = defaultdict(int)
    for elem in nums:
        quantity[elem] += 1
    return quantity

def find_max(quantity: defaultdict)-> int:
    max_val, max_key = 0, 0
    for key, val in quantity.items():
        if val > max_val:
            max_key, max_val = key, val
    return max_key

def majority_element (nums: list) -> int:
    quantity = find_count(nums)
    return find_max(quantity) 
    
def test_majority_element():
        result1 = majority_element([3, 2, 3])
        assert result1 == 3

        result1 = majority_element([2, 2, 1, 1, 1, 2, 2])
        assert result1 == 2
        
test_majority_element()
```

5. **Optional** Write an algorithm to determine if a number n is happy.

    A happy number is a number defined by the following process:

        * Starting with any positive integer, replace the number by the sum of the squares of its digits.
        * Repeat the process until the number equals 1 (where it will stay), or it loops endlessly in a cycle which does not include 1.
        * Those numbers for which this process ends in 1 are happy.

    Return true if n is a happy number, and false if not.



```python
def iteration (number: int) -> int:
    answer = 0
    while number:
        answer += (number % 10) ** 2
        number = number // 10
    return answer

def is_happy(n : int) -> bool:
    loops = set()
    while n != 1 and not (n in loops):
        loops.add (n)
        n = iteration(n)    
    return n == 1


def test_is_happy():
        result1 = is_happy(19)
        assert result1 is True

        result1 = is_happy(2)
        assert result1 is False

test_is_happy()

```

6. Given a string s, find the length of the longest substring without repeating characters.

```python
def is_unique (s: str)-> bool:
    for pos, char in enumerate(s[:-1]):
        if char in s[pos + 1:]:
            return False
    return True

def length_of_longest_substring(s: str) -> int:
    if not s:
        return 0

    max = len(s)
    length = max

    while max != 1:
        for i in range (length - max):
            if is_unique(s[i: i + max]):
                return max
        max -= 1
    
    return 1

def test_length_of_longest_substring():
        result1 = length_of_longest_substring('abcabcabc')
        assert result1 == 3

        result2 = length_of_longest_substring('bbbbb')
        assert result2 == 1

        result3 = length_of_longest_substring("pwwkew")
        assert result3 == 3

test_length_of_longest_substring()
```

7. Given two strings s and t, determine if they are isomorphic.

    Two strings s and t are isomorphic if the characters in s can be replaced to get t.

    All occurrences of a character must be replaced with another character while preserving the order of characters. No two characters may map to the same character, but a character may map to itself.

```python
from collections import defaultdict

def quantity (text: str) -> int:
        quan = defaultdict(int)
        for elem in text:
                quan[elem] += 1
        return quan

def is_isomorphic(s: str, t: str) -> bool:
        quant_s = quantity(s)
        quant_t = quantity(t)
        # return set(quant_s.values()) == set(quant_t.values())
        return list(quant_s.values()) == list(quant_t.values())

def test_is_isomorphic():
    assert is_isomorphic('egg', 'add') is True

    assert is_isomorphic("foo", "bar") is False

    assert is_isomorphic('paper', 'title') is True

    assert is_isomorphic('paper', 'titlek') is False

test_is_isomorphic()
```
